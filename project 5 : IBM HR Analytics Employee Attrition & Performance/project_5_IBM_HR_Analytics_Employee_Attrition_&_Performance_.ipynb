```python
# Drop unnecessary columns
df = df.drop(['EmployeeCount','EmployeeNumber', 'Over18', 'StandardHours'], axis = 1)
```

Convert target column â€œAttritionâ€ into numbers

Our target column (the one we want to predict) is Attrition â€” it has â€œYesâ€ and â€œNoâ€.
Letâ€™s convert it to 1 for Yes, 0 for No.

```python
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})
```

Lets Separate categorical and numeric columns

```python
categorical = df.select_dtypes(include = ['object']).columns
numerical = df.select_dtypes(exclude = ['object']).columns
```

```python
list(categorical)
```

```python
list(numerical)
```

```python
# Lets Encode categorical columns
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for col in categorical:
  df[col] = le.fit_transform(df[col])
```

```python
df.head()
```

```python
df.info()
```

Split data into input (X) and output (y)

```python
x = df.drop('Attrition', axis = 1)
y = df['Attrition']
```

```python
# Split data into training and testing sets
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y ,test_size = 0.3, random_state = 42 )      # Weâ€™ll use 70% of data for training and 30% for testing.
```

Standardize

Some columns like salary have big numbers, others are small â€” scaling helps models perform better.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

Lets Build our first model: Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

log_model = LogisticRegression()
log_model.fit(x_train, y_train)

y_pred = log_model.predict(x_test)
```

```python
# Lets Check model performance
print('Accuracy: ', accuracy_score(y_test, y_pred))
print('\n Classification Report :',  classification_report(y_test, y_pred))
print('\n Confusion Matrix :', confusion_matrix(y_test, y_pred))
```

```python
# Try another model: Random Forest

from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier()
rf_model.fit(x_train, y_train)

y_pred_rf = rf_model.predict(x_test)

print('Random Forest Accuracy Score: ', accuracy_score(y_test, y_pred_rf))
```

```python
# Compare models quickly
models = {
    'Logistic Regression' : accuracy_score(y_test, y_pred),
    'Random Forest' : accuracy_score(y_test, y_pred_rf)
}

for name, acc in models.items():
  print(f'{name} : {acc*100:.2f}%')
```

Feature Importance (Random Forest)

```python
# get feature importance from the trained random forest model
feature_importance = pd.DataFrame({
    'Feature' : x.columns,
    'Importance' : rf_model.feature_importances_
}).sort_values(by = 'Importance', ascending = False)
```

```python
feature_importance.head(10)
```

```python
# Visualize feature importance
plt.figure(figsize = (10,6))
sns.barplot(data = feature_importance.head(10), y = 'Feature', x = 'Importance', palette = 'mako')
plt.title('Top 10 Important Feature Influencing Attrition')
plt.show()
```

**Observation:**

- OverTime â€” Employees working overtime are more likely to leave.

- MonthlyIncome â€” Lower salary â†’ higher chance of leaving.

- TotalWorkingYears / YearsAtCompany â€” Employees with fewer years tend to leave.

- Age â€” Younger employees (20sâ€“30s) leave more often.

- WorkLifeBalance & JobSatisfaction â€” Poor balance or satisfaction increases attrition.

```python
# Confusion Matrix Heatmap : visualize how well the model predicted.

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred_rf)

plt.figure(figsize = (11,6))
sns.heatmap(cm, annot = True, fmt = 'd', cmap = 'Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Random Forest Confusion Matrix')
plt.show()
```

**Observation:**

This project analyzed the IBM HR Analytics dataset to understand the key factors behind employee attrition.
