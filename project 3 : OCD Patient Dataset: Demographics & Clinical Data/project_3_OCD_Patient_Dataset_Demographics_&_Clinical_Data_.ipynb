**Machine Learning**

```python
# Identify categorical columns
categorical_cols = df.select_dtypes(include = ['object']).columns.tolist()
```

```python
categorical_cols
```

```python
# Encode categorical columns using LabelEncoder
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in categorical_cols:
  df[col] = le.fit_transform(df[col].astype(str))

df.head()
```

```python
# Split into features and target
x = df.drop(columns = ['Y-BOCS Score (Obsessions)'])
y = df['Y-BOCS Score (Obsessions)']
```

```python
x.shape
```

```python
y.shape
```

```python
#  Train-test split
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
# Train a Linear Regression Model
from sklearn.linear_model import LinearRegression

# create the model
model = LinearRegression()

# fit the model on training data
model.fit(x_train, y_train)

# make predictions on test data
y_pred = model.predict(x_test)
```

```python
# Evaluate Model Performance
from sklearn.metrics import mean_absolute_error , mean_squared_error, r2_score
import numpy as np

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print('Mean Squared Error : ', mse)
print('Mean Absolute Error : ', mae)
print('Root Mean Squared Error : ', rmse)
print('R2 Score : ', r2)
```

 The Linear Regression model shows poor performance because the relationship between patient demographics and OCD obsession severity is weak and non-linear. We need to use more advanced models to capture hidden patterns.

```python
# Train Decision Tree and Random Forest Regressors

from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

# Create models
dt_model = DecisionTreeRegressor(random_state = 42)
rf_model = RandomForestRegressor(random_state= 42, n_estimators = 100)

# Train models
dt_model.fit(x_train, y_train)
rf_model.fit(x_train, y_train)

# Predictions
y_pred_dt = dt_model.predict(x_test)
y_pred_rf = rf_model.predict(x_test)
```

```python
# Evaluate both models
def evaluate_model(y_test, y_pred, model_name):
  mse = mean_squared_error(y_test, y_pred)
  mae = mean_absolute_error(y_test, y_pred)
  rmse = np.sqrt(mse)
  r2 = r2_score(y_test, y_pred)
  print(f'\n {model_name} Performance:')
  print('Mean Squared Error : ', mse)
  print('Mean Absolute Error : ', mae)
  print('Root Mean Squared Error : ', rmse)
  print('R2 Score : ', r2)
```

```python
evaluate_model(y_test, y_pred_dt, 'Decision Tree Regressor')
evaluate_model(y_test, y_pred_rf, 'Random Forest Regressor')
```

The models were unable to achieve good predictive performance, indicating that the available demographic and clinical features in the dataset do not sufficiently explain the variability in OCD obsession severity. This suggests that additional behavioral, psychological, or medical variables might be required for accurate prediction.

```python
# Feature Importance Visualization
# Even if the accuracy is low, we can still analyze which features the model considered important:

importances = rf_model.feature_importances_
features = x.columns
fi_df = pd.DataFrame({'Feature' : features, 'Importance' : importances})
fi_df = fi_df.sort_values(by = 'Importance', ascending = False)

plt.figure(figsize = (10,7))
sns.barplot(x = 'Importance', y = 'Feature', data = fi_df.head(10))
plt.title('Top 10 Feature Importance in Random Forest Regressor')
plt.show()
```

**Observation:**

- The Decision Tree model overfitted the data and performed poorly on the test set (R = 0.78).

- The Random Forest model showed a slightly better but still weak performance (R = 0.03).

- This indicates that the current features are not strong predictors of obsession severity.

- Feature importance analysis showed that variables like Family History and Duration of Symptoms contributed more, but not enough to yield high accuracy.

- Future improvements can include adding more detailed psychological or behavioral data to better capture the underlying patterns.

# **Conslusion :**

In this project, we analyzed the OCD Patient Dataset: Demographics and Clinical Data to understand how various demographic and clinical factors relate to the severity of obsessivecompulsive disorder (OCD), specifically the Y-BOCS (Obsessions) score.

We began by loading and cleaning the dataset, checking for missing values, and filling null entries in the Previous Diagnoses and Medications columns with Unknown to maintain completeness. Exploratory data analysis and visualization were performed to understand the age, gender, and Y-BOCS score distributions. The correlation heatmap revealed that most numeric features such as Age, Duration of Symptoms, and Y-BOCS (Compulsions) have very weak relationships with obsession severity.

After preprocessing and encoding categorical variables, several regression models were trained, including Linear Regression, Decision Tree Regressor, and Random Forest Regressor. However, the models achieved low performance (R near zero), indicating that the provided demographic and clinical features do not strongly explain OCD obsession severity. The Random Forest performed slightly better than other models, and its feature importance analysis highlighted variables such as Duration of Symptoms, Age, and Compulsion Type as having minor influence, while features like Patient ID and Diagnosis Date were mistakenly treated as significant due to their numeric nature.

The overall conclusion is that the current dataset lacks strong predictive features for modeling OCD severity. The severity of OCD likely depends on deeper psychological, behavioral, and medical variables not captured in this dataset. Future work should include additional factors such as therapy duration, anxiety level, lifestyle patterns, and environmental influences. This would enhance model accuracy and provide more meaningful clinical insights.

Despite the low predictive accuracy, this project successfully demonstrated the complete data science workflow  from data cleaning and exploration to model building, evaluation, and interpretation  providing valuable experience in understanding real-world healthcare datasets.
