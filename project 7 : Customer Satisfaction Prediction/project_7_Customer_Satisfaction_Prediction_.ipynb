```python
df = df[df['Customer Satisfaction Rating'].notnull()]

df = df.drop(['Resolution','Time to Resolution','First Response Time'], axis = 1)
```

```python
df.isnull().sum()
```

```python
# Convert Date of Purchase to datetime
df['Date of Purchase'] = pd.to_datetime(df['Date of Purchase'])


# Create new date features
df['Purchase_Year'] = df['Date of Purchase'].dt.year
df['Purchase_Month'] = df['Date of Purchase'].dt.month
df['Purchase_Day'] = df['Date of Purchase'].dt.day
```

```python
# Drop the original date column
df = df.drop('Date of Purchase', axis = 1)
```

```python
# Tickets Raised by Age Group

# Create age groups
bins = [0,20,30,40,50,60,70,80,90,100]
labels = ['0-20','21-30','31-40','41-50','51-60','61-70','71-80','81-90','91-100']

df['Age Group'] = pd.cut(df['Customer Age'],bins = bins, labels = labels, right = False)

# Count tickets raised by each age group
tickets_age = df['Age Group'].value_counts().sort_index()

# Lets plot
plt.figure(figsize = (10,6))
tickets_age.plot(kind = 'bar', color = 'skyblue')
plt.title('Tickets Raised by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Tickets')
plt.xticks(rotation = 45)
plt.grid(axis = 'y')
plt.show()
```

**Observation:**

The 21â€“40 age group generates the highest number of support tickets, matching the age distribution pattern. Very young and older age groups raise fewer tickets.

**Machine Learning**

```python
# Lets Encode the text column into numeric
from sklearn.preprocessing import LabelEncoder

# Create the label encoder
le = LabelEncoder()

# Encode all text columns
for col in df.select_dtypes(include='object').columns:
    df[col] = le.fit_transform(df[col])

df.head()
```

```python
# Encode Age Group (the only column left with strings)
df['Age Group'] = df['Age Group'].astype(str)
df['Age Group'] = le.fit_transform(df['Age Group'])
```

```python
df.head()
```

```python
# Split into X (features) and y (target)

x = df.drop(['Customer Satisfaction Rating'], axis = 1)
y = df['Customer Satisfaction Rating']
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split

# train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)

```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state = 42)

model.fit(x_train, y_train)

y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# accuracy
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy: ', accuracy)

# Classification Report
print('\nClassification Report: ')
print(classification_report(y_test, y_pred))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize  = (9,8))
sns.heatmap(cm, annot = True, cmap = 'Blues', fmt = 'd')
plt.title('Confusin Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

**Observation:**

The model struggles to correctly predict ratings and distributes predictions across all classes. This confirms the dataset lacks strong patterns for accurate prediction.
